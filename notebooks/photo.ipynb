```python
import cv2
import os
```

```python
video_dir = '/media/admin01/storage4/mbti_data/train_dir/train_data'
# output_dir = '/media/admin01/storage4/mbti_data/train_dir/photo_output'
output_dir = '/home/stepan2/work/photo_output'
```

```python
for i in range(1, 76):
    os.makedirs(os.path.join(output_dir, f'training80_{i:02d}_photo'), exist_ok=True)
```

```python
for i in range(1, 76):
    folder_path = os.path.join(video_dir, f'training80_{i:02d}')
    output_folder_path = os.path.join(output_dir, f'training80_{i:02d}_photo')
    
    # Проверяем, существует ли папка
    if not os.path.exists(folder_path):
        print(f"Папка {folder_path} не найдена.")
        continue

    # Перебираем все файлы в папке
    for video_file in os.listdir(folder_path):
        if video_file.endswith('.mp4'):
            video_path = os.path.join(folder_path, video_file)
            output_photo_path = os.path.join(output_folder_path, f'{os.path.splitext(video_file)[0]}_middle_frame.jpg')

            # Открываем видео
            cap = cv2.VideoCapture(video_path)
            frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

            # Переходим к середине видео
            middle_frame = frame_count // 2
            cap.set(cv2.CAP_PROP_POS_FRAMES, middle_frame)

            # Читаем кадр
            success, frame = cap.read()
            if success:
                # Сохраняем кадр как изображение JPG
                cv2.imwrite(output_photo_path, frame)
                print(f"Кадр из середины {video_path} сохранен как {output_photo_path}")
            else:
                print(f"Не удалось извлечь кадр из {video_path}")

            # Освобождаем ресурс
            cap.release()
```
